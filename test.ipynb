```python
import os

folder_path = 'op'

files_content = {}
# Iterate over the files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        # Do something with the file
        files_content[filename] = open(file_path).read()
```

```python
filesnotes = {file : [] for file in files_content}
for file in files_content:
    for token in files_content[file].split():
        if token.startswith('NOTE_ON'):
            filesnotes[file].append(int(token.split('=')[1]))
```

```python
import numpy as np
counter21 = 0
counter12 = 0
for file in filesnotes:
    for note in filesnotes[file]:
        if note < 21:
            counter21 += 1
            break
    for note in filesnotes[file]:
        if note < 12:
            print(file)

            counter12 += 1
            break
        
print(counter21, counter12)
   
```
